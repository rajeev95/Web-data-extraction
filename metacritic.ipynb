```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#containing in a source variable
src = open('metacritic.txt','r').read()
```

```python
#creating the beautiful soup object 
soup = BeautifulSoup(src)
```

```python
k=soup.find('table',{'class':'credits company_credits'}).tbody
```

```python
s=[v.text for v in k.find_all('td')]
```

```python
listt=[]
x=0
y=3
while (x<len(s)):
    listt.append(s[x:y])
    x=x+3
    y=y+3
```

```python
list_1=[]
for k in range(0,30):
    l=str(listt[k]).split('\\n')[1]
    list_1.append(l)
```

```python
list_2=[]
for k in range(0,30):
    l=str(listt[k]).split('\\n')[2]
    list_2.append(l)
```

```python
list_3=[]
for k in range(0,30):
    l=str(listt[k]).split('\\n')[6]
    list_3.append(l)
```

```python
metacritic = pd.DataFrame({'Metascore':list_1,'Release Title':list_2,'Userscore':list_3})
```

```python
df= pd.read_csv('Franchise_MCU.csv')
```

```python
df= pd.merge(df,metacritic,how='left',on='Release Title')
```

```python
df.fillna('-',inplace=True)
```
